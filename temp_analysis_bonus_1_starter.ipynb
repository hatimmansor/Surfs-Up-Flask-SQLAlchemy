### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months
df
```

```python
# Identify the average temperature for June
june_df=df.loc[df.index.month==7].groupby('station')['tobs'].mean().to_frame()
june_df
```

```python
# Identify the average temperature for December
december_df=df.loc[df.index.month==12].groupby('station')['tobs'].mean().to_frame()
december_df
```

```python
# Create collections of temperature data
```

```python
# Run paired t-test
stats.ttest_rel(june_df['tobs'],december_df['tobs'])
```

### Analysis

```python
the p-value of the test is smaller than 0.00002 which is less than the significance level alpha (e.g., 0.05). This means that we can draw the conclusion that the everage temperature has decreased in december than june 
```
